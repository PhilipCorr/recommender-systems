### Popularity Based Recommenders

```python
frame = pd.read_csv("rating_final.csv")
cuisine = pd.read_csv("chefmozcuisine.csv")
geodata = pd.read_csv("geoplaces2.csv", encoding="latin1")
```

```python
frame.head()
```

```python
geodata.head()
```

```python
rating_count = pd.DataFrame(frame.groupby("placeID")["rating"].count())
ordered_rating = rating_count.sort_values('rating', ascending=False)
ordered_rating.head()
```

```python
top_ratings = ordered_rating.index.values.tolist()[0:5]
```

```python
top_rated_places = pd.DataFrame(top_ratings, columns=['placeID'])
```

```python
summary = pd.merge(top_rated_places, cuisine, on='placeID')
```
